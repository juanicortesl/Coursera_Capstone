```python
import pandas as pd
import numpy as np
```

```python
# we obtain and pre process df like stated in part 1 of the assignment
df = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')[0]
df.drop(df.loc[df['Borough']=="Not assigned"].index, inplace=True)
df = (df.groupby(['Postal Code', 'Borough'])['Neighbourhood'].agg([('Neighbourhood', ', '.join)]).reset_index())
df = df.rename({'Postal Code':'PostalCode'}, axis=1)
```

```python
# I couldnt access the coordinates api, so I just downloaded the csv file from  http://cocl.us/Geospatial_data
coords = pd.read_csv("Geospatial_Coordinates.csv")
coords = coords.astype("object")
coords = coords.rename({'Postal Code':'PostalCode'}, axis=1)
coords.head()
```

```python
#merging first data frame with coordinates df
df = df.merge(coords)
df.head(10)
```
